```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import geopandas as gp
import shapely
import fiona
import requests
import re
import folium

from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from folium import plugins
from folium.plugins import HeatMap
from bs4 import BeautifulSoup
from datetime import datetime
```

```python

if os.path.isfile("atlcrime.csv"):
    filepath = "atlcrime.csv"
    print("loading from file")
else:
    filepath = "https://data.world/bryantahb/crime-in-atlanta-2009-2017/workspace/file?filename=atlcrime.csv"

    print("loading from the internet")

crime_data = pd.read_csv(filepath, parse_dates=[3], dayfirst=False)
print("done")
```

```python
map1 = folium.Map(width=500, height=500, location=[33.77101, -84.38895], zoom_start=11, min_zoom=8)

for i in range (len(crime_data)):
    if (crime_data.iloc[i].crime == "HOMICIDE"):
        folium.Marker(location=[crime_data.iloc[i].lat, 
                        crime_data.iloc[i].long], 
                        popup=crime_data.iloc[i].date).add_to(map1)

homicide = ["HOMICIDE"]
homicide_list = crime_data[[x in homicide for x in crime_data.crime]]

lat = homicide_list.lat.tolist()
lng = homicide_list.long.tolist() #this is based on all crimes create a subset of just homicide crimes.

HeatMap(list(zip(lat, lng))).add_to(map1)

folium.Marker(location=[33.75431,-84.372149], popup="21 Savage Neighbourhood", icon=folium.Icon(color="orange")).add_to(map1)
```
